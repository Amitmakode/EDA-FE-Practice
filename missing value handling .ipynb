# Filling the null values

```python
df0 = data.fillna(value= 22)    #fill values with a spacifice value
df0
```

```python
df1 = data.fillna(method= "pad")    #fill value by previous value in vertically
df1
```

```python
df2 = data.fillna(method= "bfill")    #fill nan value by next value vertically 
df2
```

```python
df3 = data.fillna(method= "pad", axis = 1)   #fill value by using previous value horizontaly
df3
```

```python
df4 = data.fillna(method= "bfill", axis = 1)    #fill value by using next value horizontaly
df4
```

```python
df5 = data.fillna({"society":"bang-bang", "balcony":"10"})  # fill diffrent values in different column
df5
```

```python
df6 = data.fillna(value = data['balcony'].mean()) #fill values with mean, median, mode, max, min, 
df6
```

# Drop NaN

```python
df7 = data.dropna(how = "any")  #drop nan value by using min one nan value present in the row
df7
```

```python
df8 = data.dropna(how="all")    # drop nan value by using all value present in the row is nan
df8
```

# replace null values

```python
df9 = data.replace(to_replace=np.nan, value = 529)   #replace value by using some diffrent value
df9
```

```python
df10 = data.replace(to_replace=3.0, value = 5.0)     #replace exsisting value 
df10
```
